# 현상 5
- 컨텐츠 생산은 압도적으로 미국이 많고, 그 뒤로는 인도, 영국, 캐나다, 프랑스, 일본 순이다. 
- 컨텐츠 생산량은 넷플릭스 서비스 시작 시기와 인구수에 비례한 것으로 생각된다.   
- 중국은 예외적으로 이에 비해 컨텐츠 생산량이 현저히 적은 것을 볼 수 있는데, 중국에서는 넷플릭스 사용이 금지되어 있어 컨텐츠에 관여하기 어려운 것이 원인으로 생각된다.  

```python
from sklearn.preprocessing import MultiLabelBinarizer # Similar to One-Hot Encoding

def relation_heatmap(df, title):
    df['genre'] = df['listed_in'].apply(lambda x :  x.replace(' ,',',').replace(', ',',').split(',')) 
    Types = []
    for i in df['genre']: Types += i
    Types = set(Types)
    print(f"There are {len(Types)} types in the Netflix {title} Dataset")    
    test = df['genre']
    mlb = MultiLabelBinarizer()
    res = pd.DataFrame(mlb.fit_transform(test), columns=mlb.classes_, index=test.index)
    corr = res.corr()
    mask = np.zeros_like(corr, dtype=np.bool)
    mask[np.triu_indices_from(mask)] = True
    fig, ax = plt.subplots(figsize=(15, 14))
    pl = sns.heatmap(corr, mask=mask, cmap= "coolwarm", vmax=.5, vmin=-.5, center=0, square=True, linewidths=.7, cbar_kws={"shrink": 0.6})
    
    plt.show()
```

```python
relation_heatmap(data[data['type']=='Movie'], 'Movie')
```

# 현상 6
- 국제,독립 영화 장르와 드라마 장르는 높은 양의 상관관계를 보이며, 이는 국제, 독립 영화가 대부분 기승전결의 스토리가 있는 극형식이기 때문인 것으로 생각된다.  
- 드라마 장르와 다큐멘터리 장르는 높은 음의 상관관계를 보이며, 다큐멘터리 장르는 대부분 스토리가 있는 극형식이 아닌 정보 전달 위주이기 때문인 것으로 생각된다.

```python
relation_heatmap(data[data['type']=='TV Show'], 'TV Show')
```
