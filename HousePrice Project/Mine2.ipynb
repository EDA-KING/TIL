```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
raw_train = train.copy()
raw_test = test.copy()
```

```python
alldf = pd.concat([train,test],ignore_index=True)
```

```python
alldf.shape
```

```python
null_dict = alldf.isnull().sum()
null_dict2 = {}
for col, null_count in null_dict.items():
    if null_count > 0:
        null_dict2[col] = null_count
```

```python
null_dict2
```

```python
alldf.shape
```

```python
for col, null_count in null_dict2.items():
    if col == 'SalePrice':
        continue
    if null_count >= 1000:
        alldf.drop(col,axis=1,inplace=True)
```

```python
alldf.shape
```

```python
null_dict = alldf.isnull().sum()
null_dict2 = {}
for col, null_count in null_dict.items():
    if null_count > 0:
        null_dict2[col] = null_count
```

```python
null_dict2
```

```python
str_col = []
num_col = []
for col in null_dict2:
    if alldf[col][0].__class__.__name__ == 'str':
        str_col.append(col)
    else:
        num_col.append(col)
```

```python
str_col
```

```python
num_col
```

```python
for col in str_col:
    alldf[col].fillna( alldf[col].mode()[0],inplace=True)
```

```python
for col in num_col:
    alldf[col].fillna( alldf[col].median(),inplace=True)
```

```python
alldf
```

```python
str_col = []
for col in alldf.columns:
    if alldf[col][0].__class__.__name__ == 'str':
        str_col.append(col)
```

```python
alldf
```

```python
str_df = pd.DataFrame(alldf[str_col], columns=str_col )
alldf = pd.concat( [alldf,pd.get_dummies(str_df)],axis=1)
alldf
```

```python
alldf.drop(str_col,axis=1,inplace=True)
alldf
```

```python
from sklearn.linear_model import LinearRegression
```

```python
train = alldf.iloc[:raw_train.shape[0],:] 
test = alldf.iloc[raw_train.shape[0]:,:]
```

```python
test.drop('SalePrice',axis=1,inplace=True)
```

```python
x_train = train.iloc[:,:-1]
y_train = train['SalePrice']
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
y_pred = model.predict(test)
```

```python
y_pred
```

```python
submission = pd.DataFrame({'Id':raw_test['Id'],
                          'SalePrice':y_pred})
```

```python
submission.to_csv('submission_Mine2.csv',index=False)
```
