```python
for col in train.columns:
    if (train[col][0].__class__.__name__ == 'str') | (train[col][0].__class__.__name__ == 'float'):
        train.drop(col,axis=1,inplace=True)  
```

```python
null_col = []
for col in train.columns:
    if train[col].isnull().any():
        null_col.append(col)

null_col
```

```python
for col in null_col:
    train[col].fillna(train[col].median(),inplace=True)
```

```python
null_col = []
for col in test.columns:
    if test[col].isnull().any():
        null_col.append(col)

null_col
```
