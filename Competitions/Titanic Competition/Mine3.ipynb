```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

raw_train = train.copy()
raw_test = test.copy()
```

```python
allData = pd.concat([train,test],ignore_index=True)
```

```python
allData
```

```python
allData.drop(['PassengerId','Ticket','Cabin'],axis=1,inplace=True)
```

```python
allData.head()
```

```python
allData['Title'] = allData['Name'].str.extract( '([a-zA-Z]+)\.' )
```

```python
title_dict = allData['Title'].value_counts()
```

```python
rare_titles = []
for title, count in title_dict.items():
    if count < 10:
        rare_titles.append(title)
```

```python
rare_titles
```

```python
allData['Title'] = allData['Title'].replace(rare_titles,'Rare')
```

```python
allData['Title'].value_counts()
```

```python
allData.head()
```

```python
allData.drop('Name',axis=1,inplace=True)
```

```python
allData.head()
```

```python
mean = allData['Age'].mean()
std = allData['Age'].std()
null_count = allData['Age'].isnull().sum()

pred_ages = np.random.randint(mean - std, mean + std , null_count)

allData.loc[allData['Age'].isnull(),'Age'] = pred_ages
```

```python
allData.isnull().sum()
```

```python
allData['Fare'].fillna(allData['Fare'].median(),inplace=True)
allData['Embarked'].fillna( allData['Embarked'].mode()[0],inplace=True)
```

```python
allData.isnull().sum()
```

```python
allData.head()
```

```python
allData['AgeBand'] = pd.cut(allData['Age'],4)
allData['FareBand'] = pd.qcut(allData['Fare'],4)
```

```python
allData.groupby('AgeBand')[['Survived']].mean()
```

```python
allData.groupby('FareBand')[['Survived']].mean()
```

```python
def AgetoOrdinal(age):
    if age <= 20:
        return 1
    if age <= 40:
        return 2
    if age <= 60:
        return 3
    else:
        return 4
```

```python
def FaretoOrdinal(fare):
    if fare <= 7.89:
        return 1
    if fare <= 14.45:
        return 2
    if fare <= 31.27:
        return 3
    else:
        return 4
```

```python
allData['Age'] = allData['Age'].map( AgetoOrdinal )
allData['Fare'] = allData['Fare'].map( FaretoOrdinal )
```

```python
allData.drop(['AgeBand','FareBand'],axis=1,inplace=True)
```

```python
allData.head()
```

```python
allData['FamilySize'] = allData['SibSp'] + allData['Parch']
```

```python
allData.drop(['SibSp','Parch'],axis=1,inplace=True)
```

```python
allData.head()
```

```python
df = pd.DataFrame( {'Sex':allData['Sex'],
                   'Embarked':allData['Embarked'],
                   'Title':allData['Title']} )
df
```

```python
dummy_df = pd.get_dummies( df )
```

```python
allData = pd.concat([allData,dummy_df],axis=1)
```

```python
allData.head()
```

```python
allData.drop(['Sex','Embarked','Title'],axis=1,inplace=True)
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.metrics import accuracy_score, log_loss
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression

models = [
    KNeighborsClassifier(3),
    SVC(probability=True),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
	AdaBoostClassifier(),
    GradientBoostingClassifier(),
    GaussianNB(),
    LinearDiscriminantAnalysis(),
    QuadraticDiscriminantAnalysis(),
    LogisticRegression()]
```

```python
train = allData.iloc[:train.shape[0],:] 
test = allData.iloc[train.shape[0]:,:]
```

```python
x_train = train.iloc[:,1:] 
y_train = train.iloc[:,0]
```
